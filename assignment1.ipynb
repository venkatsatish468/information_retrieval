```python
import nltk
nltk.download("punkt")
nltk.download('stopwords')
import string
import re
import os
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
def removePunctuation(text):
    text = text.translate(str.maketrans(string.punctuation, ' '*len(string.punctuation)))
    return "".join([char for char in text if char not in string.punctuation])

def tokens(text):
    return nltk.word_tokenize(text)

def removeNumbers(oldList):
    file_remove_num = []
    for word in oldList:
        new_word = re.sub(r'\d+','',word)
        if new_word != '':
            file_remove_num.append(new_word)
    return file_remove_num

def removeStopWords(numList):
    stop_words = stopwords.words('english')
    file_remove_stop_words = [word for word in numList if word not in stop_words]
    return file_remove_stop_words

def stemming(stopList):
    porter = PorterStemmer()
    file_stemmed = [porter.stem(word) for word in stopList]
    return file_stemmed

def toCombine(stemList):
    file_combined=" ".join(stemList)
    return file_combined

def toSave(text,fname):
    fw=open("C:/Users/satish/Documents/ir_assignment/assignment1/"+fname,"w")
    fw.write(text)
    fw.close()
    
```

```python
file1=open("C:/Users/satish/Documents/ir_assignment/assignment1/Text1.txt","r")
file2=open("C:/Users/satish/Documents/ir_assignment/assignment1/Text2.txt","r")
file3=open("C:/Users/satish/Documents/ir_assignment/assignment1/Text3.txt","r")
fileLarge=open("C:/Users/satish/Documents/ir_assignment/assignment1/LargeText.txt","r")
file1Text=file1.read().lower()
file2Text=file2.read().lower()
file3Text=file3.read().lower()
largeFileText=fileLarge.read().lower()
file1.close()
file2.close()
file3.close()
fileLarge.close()
print(file1Text)
```

```python
file1RemovePunctuation=removePunctuation(file1Text)
file2RemovePunctuation=removePunctuation(file2Text)
file3RemovePunctuation=removePunctuation(file3Text)
largeFileRemovePunctuation=removePunctuation(largeFileText)
print(file1RemovePunctuation)
```

```python
file1Tokens=tokens(file1RemovePunctuation)
file2Tokens=tokens(file2RemovePunctuation)
file3Tokens=tokens(file3RemovePunctuation)
largeFileTokens=tokens(largeFileRemovePunctuation)
print("Length After generating Tokens :",len(file1Tokens))
print(" ")
print(file1Tokens)
```

```python
file1RemoveNumbers=removeNumbers(file1Tokens)
file2RemoveNumbers=removeNumbers(file2Tokens)
file3RemoveNumbers=removeNumbers(file3Tokens)
largeFileRemoveNumbers=removeNumbers(largeFileTokens)
print("Length after removing numbers :",len(file1RemoveNumbers))
print(" ")
print(file1RemoveNumbers)
```

```python
file1RemoveStopWords=removeStopWords(file1RemoveNumbers)
file2RemoveStopWords=removeStopWords(file2RemoveNumbers)
file3RemoveStopWords=removeStopWords(file3RemoveNumbers)
largeFileRemoveStopWords=removeStopWords(largeFileRemoveNumbers)
print("Length after Removing Stopwords : ",len(file1RemoveStopWords))
print(" ")
print(file1RemoveStopWords)
```

```python
file1Stemming=stemming(file1RemoveStopWords)
file2Stemming=stemming(file2RemoveStopWords)
file3Stemming=stemming(file3RemoveStopWords)
largeFileStemming=stemming(largeFileRemoveStopWords)
print("length after Stemming :",len(file1Stemming))
print(" ")
print(file1Stemming)
```

```python
file1CombinedLexi=toCombine(file1RemoveNumbers)
file2CombinedLexi=toCombine(file2RemoveNumbers)
file3CombinedLexi=toCombine(file3RemoveNumbers)
largeFileCombinedLexi=toCombine(largeFileRemoveNumbers)
print(file1CombinedLexi)
```

```python
file1CombinedStop=toCombine(file1RemoveStopWords)
file2CombinedStop=toCombine(file2RemoveStopWords)
file3CombinedStop=toCombine(file3RemoveStopWords)
largeFileCombinedStop=toCombine(largeFileRemoveStopWords)
print(file1CombinedStop)
```

```python
file1CombinedStem=toCombine(file1Stemming)
file2CombinedStem=toCombine(file2Stemming)
file3CombinedStem=toCombine(file3Stemming)
largeFileCombinedStem=toCombine(largeFileStemming)
print(file1CombinedStem)
```

```python
toSave(file1CombinedLexi,"file1AfterLexical.txt")
toSave(file1CombinedStop,"file1AfterStopWords.txt")
toSave(file1CombinedStem,"file1AfterStemming.txt")

toSave(file2CombinedLexi,"file2AfterLexical.txt")
toSave(file2CombinedStop,"file2AfterStopWords.txt")
toSave(file2CombinedStem,"file2AfterStemming.txt")

toSave(file3CombinedLexi,"file3AfterLexical.txt")
toSave(file3CombinedStop,"file3AfterStopWords.txt")
toSave(file3CombinedStem,"file3AfterStemming.txt")

toSave(largeFileCombinedLexi,"largeAfterLexical.txt")
toSave(largeFileCombinedStop,"largeAfterStopWords.txt")
toSave(largeFileCombinedStem,"largeAfterStemming.txt")
```
