```python
import nltk
nltk.download("punkt")
nltk.download('stopwords')
import string
import re
import os
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import pandas as pd
import math
```

```python
def removePunctuation(text):
    return "".join([char for char in text if char not in string.punctuation])

def createTokens(text):
    return nltk.word_tokenize(text)

def removeNumbers(oldList):
    file_remove_num = []
    for word in oldList:
        new_word = re.sub(r'\d+','',word)
        if new_word != '':
            file_remove_num.append(new_word)
    return file_remove_num

def removeStopWords(numList):
    stop_words = stopwords.words('english')
    file_remove_stop_words = [word for word in numList if word not in stop_words]
    return file_remove_stop_words

def stemming(stopList):
    porter = PorterStemmer()
    file_stemmed = [porter.stem(word) for word in stopList]
    return file_stemmed

def preprocessData(String):
    String = String.translate(str.maketrans(string.punctuation, ' '*len(string.punctuation)))
    queryRemovePun=removePunctuation(String)
    tokens=createTokens(queryRemovePun)
    removeNum=removeNumbers(tokens)
    RemoveStopWords=removeStopWords(removeNum)
    Stemming=stemming(RemoveStopWords)
    return Stemming

def createInvertedIndex(processedData,vocabulary):
    invertedIndex={}
    for word in vocabulary:
        invertedIndex[word]={}
        for document,tokens in processedData.items():
            invertedIndex[word][document]=tokens.count(word)
    return invertedIndex
```

### Term frequency = Number of instances the word has occured in document
### Inverse document frequency = The number of documents in which a term appears
#### IDF = log(N/(No. of documents in which the term has occured))
### Length of document = Sum of squares TF-IDF weight of all index terms in that document
### Similarity values = (Dot product of TF-IDF weights of document and query) / (length of document * length of query)

```python
def tf(df):
    for column in df:
        m = max(df[column])
        if m!=0:
            df[column] = df[column]/m
    return df

def get_idf(df):
    idf_score = {}
    N = df.shape[1]
    all_words = df.index
    word_count = df.astype(bool).sum(axis=1)
    for word in all_words:
        idf_score[word] = math.log10(N/word_count[word])
    return idf_score

def get_tfidf(data, idf_score):
    scores = {}
    for key,value in data.items():
        scores[key] = data[key]
    for doc,tf_scores in scores.items():
        for token, score in tf_scores.items():
            tf = score
            idf = idf_score[token]
            tf_scores[token] = tf * idf
    return scores

def tf_query(query_words):
    all_words = invertedIndexDf.index
    index = {}
    index["query"] = {}
    for word in all_words:
        index["query"][word] = 0
    for qword in query_words:
        if qword in all_words:
            index["query"][qword] = query_words.count(qword)
    return index

def rankDocuments(tf_idf_docs,tf_idf_query):
    query_docs = {}
    query_docs["query"] = {}
    query_length = math.sqrt(sum(tf_idf_query.loc[value] ** 2 for value in tf_idf_query.index))
    
    if(query_length==0):
        print("no match in document")
        for column in tf_idf_docs:
            query_docs["query"][column] = 0
        return query_docs
    
    for column in tf_idf_docs:
        
        num = 0
        sum_of_squares = 0
        
        for value in tf_idf_docs.index :
            sum_of_squares+=tf_idf_docs[column].loc[value] ** 2
            num+= tf_idf_docs[column].loc[value] * tf_idf_query["query"].loc[value] 
            
        doc_len = math.sqrt(sum_of_squares)
        cosine_sim = num/(doc_len*query_length)
        query_docs["query"][column] = cosine_sim
    
    return query_docs
```

```python
normalized_tf = tf(invertedIndexDf)
idf_score = get_idf(normalized_tf)
tf_idf_docs = get_tfidf(normalized_tf,idf_score)
tf_idf_docs = pd.DataFrame(tf_idf_docs)
tf_idf_docs.head()
```
